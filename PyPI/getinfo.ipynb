```python
import pickle
import subprocess
import json

import requests
from bs4 import BeautifulSoup
```

```python
all_package_name = requests.get('https://pypi.org/simple')
soup = BeautifulSoup(all_package_name.text, 'html.parser')
all_packages = soup.find_all('a')
```

```python
with open('all_package.txt', 'w') as f:
    print(all_package_name.text, file=f)
```

```python
all_package_text = [i.text for i in all_packages]
with open('all_package_names.pkl', 'wb') as f:
    pickle.dump(all_package_text, f, pickle.HIGHEST_PROTOCOL)
```

```python
with open('all_package_names.pkl', 'rb') as f:
    all_packages = pickle.load(f)
```

```python
len(all_packages)
```

```python
all_packages[1]
```

```python
subprocess.run(['pip', 'install', 'pipenv'], shell=False)
```

```python
ok = set()

for i in range(len(all_packages) // 100):
    packages = all_packages[i:i+100]
    packages = [i for i in packages if i not in ok]
    p = subprocess.run(['pipenv', 'install'] + packages, shell=False,
                       capture_output=True, text=True)
    if p.returncode != 0: 
        with open('install_log.txt', 'w+') as log:
            print(packages, p.stderr.splitlines()[-9:], sep='\n\n', file=log)
    else:
        tree = subprocess.run(['pipenv', 'graph', '--json'], shell=False,
                              capture_output=True, text=True)
        with open('dependency.tsv', 'w+') as f:
            j = json.loads(tree.stdout)
            for package in j:
                k = package['package']['key'] + '\t' + package['package']['installed_version']
                if k in ok:
                    continue
                package_into = '\t'.join(package['package'].values())
                dependency_info = '\t'.join(['\t'.join(de.values()) for de in package['dependencies']])
                if dependency_info:
                    print(package_info, dependency_info, sep='\t', file=f)
                else:
                    print(package_info, file=f)
                ok.add(k)
    
```
